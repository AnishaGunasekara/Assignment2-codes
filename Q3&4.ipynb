```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
im_og= cv.imread('/content/drive/MyDrive/Image_Processing/cf.jpg')
im_og= cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

im_1 = cv.Canny(im_og, 550, 690)
y, x = np.nonzero(im_1)

# Fitting a line to all edge points using least squares method
# Fitting a line of the form y = mx + c
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

#Plotting scatter plot of edge points and overlayinh the fitted line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', s=5)
plt.plot(x, m*x + c, color='blue')
plt.title('Edge Points and Least Squares Fit Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  # Inverting y-axis to match image coordinates
plt.grid(True)
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
im_og= cv.imread('/content/drive/MyDrive/Image_Processing/cf.jpg')
im_og= cv.cvtColor(im_og, cv.COLOR_BGR2GRAY)

im_1 = cv.Canny(im_og, 550, 690)
y, x = np.nonzero(im_1)

# Fitting a line to all edge points using least squares method
# Fitting a line of the form y = mx + c
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

slope = m
angle_radians = np.arctan(slope) # Calculate the angle in radians using arctangent function

# Convert radians to degrees
angle_degrees = np.degrees(angle_radians)

# Output the estimated angle
print("Estimated crop field angle (in degrees):", angle_degrees)

```
